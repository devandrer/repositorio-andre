```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dias = np.arange(1, 31)
vendas = np.random.randint(50, 150, size=30)
```

```python
df_vendas = pd.DataFrame({'Dia': dias,'Vendas': vendas})
```

```python
df_vendas
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df_vendas['Dia'], df_vendas['Vendas'], marker = 'o', linestyle = '-', color = 'b')
```

```python
plt.title('Acompanhamento de Vendas Di√°rias', fontsize = 16)
plt.xlabel('Dia do M√™s', fontsize = 14)
plt.ylabel('Vendas Di√°rias', fontsize = 14)
```

```python
plt.grid(True)
plt.show()
```
