```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
n_samples = 50
```

```python

data = {
    'Data': pd.date_range(start='2023-01-01', periods=n_samples, freq='D'),
    'Vendas': np.random.randint(50, 200, size=n_samples),
    'Lucro': np.random.randint(10, 50, size=n_samples)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Margem'] = df['Lucro']/df['Vendas']
```

```python
df['Margem'].min()
```

```python
df['Margem'].max()
```

```python
plt.figure(figsize=(13, 3))
plt.plot(df['Data'], df['Vendas'], marker = 'o', linestyle = '-', color = 'b')
```
